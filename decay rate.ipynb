```python
import numpy as np

# Half-life ranges in weeks
half_life_tv = (2 + 6) / 2
half_life_radio = (1 + 5) / 2
half_life_newspaper = (2 + 3) / 2
half_life_print_magazines = (4 + 8) / 2
half_life_online_upper = (2 + 4) / 2
half_life_online_lower = (1 + 2) / 2

# Calculate the decay rates
decay_rate_tv = 0.5 ** (1 / half_life_tv)
decay_rate_radio = 0.5 ** (1 / half_life_radio)
decay_rate_newspaper = 0.5 ** (1 / half_life_newspaper)
decay_rate_print_magazines = 0.5 ** (1 / half_life_print_magazines)
decay_rate_online_upper = 0.5 ** (1 / half_life_online_upper)
decay_rate_online_lower = 0.5 ** (1 / half_life_online_lower)

decay_rates = {
    "TV": decay_rate_tv,
    "Radio": decay_rate_radio,
    "Newspaper": decay_rate_newspaper,
    "Print Magazines": decay_rate_print_magazines,
    "Online (Upper Funnel)": decay_rate_online_upper,
    "Online (Lower Funnel)": decay_rate_online_lower
}

decay_rates
```

```python
def calculate_adstock(spend, alpha):
    adstock = np.zeros_like(spend)
    adstock[0] = spend[0]
    for t in range(1, len(spend)):
        adstock[t] = spend[t] + alpha * adstock[t-1]
    return adstock


for i in range(1, 8):
    channel = f'spend_channel_{i}'
    print(channel,df[channel].values)
   
```
